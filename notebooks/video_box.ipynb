```python
from ultralytics import YOLO
from moviepy.editor import VideoFileClip, CompositeAudioClip, AudioFileClip
import os

model = YOLO("../app/src/face_detection/model/yolov8n-face.pt")
```

```python
def draw_face_box_on_video(input_path):

    temp_file_name = input_path.split("/")[-2]

    VideoFileClip(input_path).audio.write_audiofile(
        f"./{temp_file_name}.mp3"
    )

    model.predict(
        source=input_path,
        save=True,
        vid_stride=1,
        save_crop=False,
        project="face_detection_results",
        name=temp_file_name,
    )

    video_with_box = VideoFileClip(f"./face_detection_results/{temp_file_name}/video.avi")
    new_audio_clip = AudioFileClip(f"./{temp_file_name}.mp3")

    video_with_box.audio = CompositeAudioClip([new_audio_clip])

    input_path = input_path.replace("video", "video_with_face_box")
    video_with_box.write_videofile(input_path)

    new_audio_clip.close()
    video_with_box.close()

    os.remove(f"./{temp_file_name}.mp3")
    os.remove(f"./face_detection_results/{temp_file_name}/video.avi")
    os.rmdir(f"./face_detection_results/{temp_file_name}")


input_path = "../app/uploads/2185f57c-e40d-4dc6-95d1-d6c84050cb15/video.mp4"
input_path = "../app/uploads/c8dd5351-0653-4929-a177-f8a23892d88d/video.mp4"

draw_face_box_on_video(input_path)
```
